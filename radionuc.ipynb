```python
import pandas as pd
import matplotlib.pyplot as plt   
import numpy as np 
import seaborn as sns
csv_file_path1 = 'dati/composite/C11_entering_particles.csv'

csv_file_path2 = 'dati/composite/O15_entering_particles.csv'

df_C11 = pd.read_csv(csv_file_path1, delimiter=',', header=None, names=['Energy', 'X', 'Y', 'Z','evt'])
df_O15 = pd.read_csv(csv_file_path2, delimiter=',', header=None, names=['Energy', 'X', 'Y', 'Z','evt'])


plt.scatter(df_C11['X'], df_C11['Y'], label='C11',)
plt.xlim(-150, 150)
plt.ylim(-100, 100)
plt.show()
```

```python
plt.hist2d(df_C11['X'], df_C11['Y'], bins=(30,30), cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(x=df_C11['X'], y=df_C11['Y'], cmap='viridis', fill=True, thresh=0, levels=100,common_grid=True,) 
plt.title('C11')
#plt.xlim(-150,150)
#plt.ylim(-100,100)
plt.show()
```
