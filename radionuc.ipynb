```python
import pandas as pd
import matplotlib.pyplot as plt   
import numpy as np 
import seaborn as sns
csv_file_path1 = 'dati/radionucbigphantom/60MeV/C11_entering_particles.csv'

csv_file_path2 = 'dati/radionucbigphantom/60MeV/O15_entering_particles.csv'

df_C11 = pd.read_csv(csv_file_path1, delimiter=',', header=None, names=['Energy', 'X', 'Y', 'Z','evt'])
df_O15 = pd.read_csv(csv_file_path2, delimiter=',', header=None, names=['Energy', 'X', 'Y', 'Z','evt'])


plt.scatter(df_C11['X'], df_C11['Y'], label='C11',)
#plt.xlim(-150, 150)
#plt.ylim(-100, 100)
plt.show()
```

```python
print(np.shape(df_C11['X']))
```

```python
plt.hist2d(df_C11['X'], df_C11['Y'], bins=(30,31), cmap='viridis')
plt.colorbar()


plt.show()
```

```python
hist2d, x_edges, y_edges = np.histogram2d(df_C11['X'], df_C11['Y'], bins=(200, 1), range=[[-300, 300], [-100, 100]])

slice_index = 0

# Extract the slice from the 2D histogram
hist_slice = hist2d[:, slice_index]

bin_centers_x = (x_edges[:-1] + x_edges[1:]) / 2

plt.hist(bin_centers_x, weights=hist_slice, color='blue', bins=10,histtype='step')

# Add labels and title
plt.xlabel('X[mm]')
plt.ylabel('Frequency')
plt.title('Densit C11')

# Show the plot
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(x=df_C11['X'], y=df_C11['Y'], cmap='viridis', fill=True, thresh=0, levels=100,common_grid=True,) 
plt.title('C11')
#plt.xlim(-150,150)
#plt.ylim(-100,100)
plt.show()
```

```python
csv_file_path3 = 'build/gamma_entering_particles.csv'


df_gamma = pd.read_csv(csv_file_path3, delimiter=',', header=None, names=['Energy', 'X', 'Y', 'Z','pX', 'pY', 'pZ','evt'])


plt.scatter(df_gamma['X'], df_gamma['Y'], label='C11',)
#plt.xlim(-150, 150)
#plt.ylim(-100, 100)
plt.show()
```

```python
plt.hist2d(df_gamma['X'], df_gamma['Y'], bins=(100,100), cmap='viridis',range=[[-50, 150], [-60, 60]])
plt.colorbar()
plt.show()
```

```python
hist2d, x_edges, y_edges = np.histogram2d(df_gamma['X'], df_gamma['Y'], bins=(100, 1), range=[[-149.99, 149.99], [-100, 100]])

slice_index = 0

# Extract the slice from the 2D histogram
hist_slice = hist2d[:, slice_index]

bin_centers_x = (x_edges[:-1] + x_edges[1:]) / 2

plt.hist(bin_centers_x, weights=hist_slice, color='blue', bins=100,histtype='step')

# Add labels and title
plt.xlabel('X[mm]')
plt.ylabel('Frequency')
plt.title('Densit gamma')

# Show the plot
plt.show()
```

```python
theta=np.arccos(df_gamma['pX']/np.sqrt(df_gamma['pX']**2+df_gamma['pY']**2+df_gamma['pZ']**2))

plt.hist(theta, bins=200, histtype='step')
plt.show()
```

```python
df_gamma['theta']=theta

df_gamma=df_gamma[np.isclose(df_gamma['theta'],1.5,atol=0.2)]

```

```python
plt.hist2d(df_gamma['X'], df_gamma['Y'], bins=(100,100), cmap='viridis',range=[[-150, 150], [-60, 60]])
plt.colorbar()
plt.show()
```
